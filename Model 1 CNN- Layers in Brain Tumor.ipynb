**Data Augmentation**

```python
##### Data Augmentation methods

Glioma_Mri = Route_of_Image_glioma+"/Te-gl_0010.jpg"
Glioma_Mri = Brain_tumorIGo.imread(Glioma_Mri)

# 1. Flipping
print("Images flipping")
Brain_tumor_hor= Brain_tumoroia.Fliplr(p=1.0)
Brain_tumor_hor_i= Brain_tumor_hor.augment_image(Glioma_Mri)
Brain_tumor_ve= Brain_tumoroia.Flipud(p=1.0)
Brain_tumor_ve_i= Brain_tumor_ve.augment_image(Glioma_Mri)
Naming=[Glioma_Mri, Brain_tumor_hor_i, Brain_tumor_ve_i]
Naming_lb = ['Brain_tumor_Ori', 'Brain_tumor_Hori flipped', 'Brain_tumor_Vert flipped']
Brain_tumor_piy.plot_images(Naming,labels=Naming_lb,img_width=180)

# 2. Rotation
print("Images Rotation")

rot_Brain_tumor = Brain_tumoroia.Affine(rotate=(-50,20))
rot_Brain_tumor_i = rot_Brain_tumor.augment_image(Glioma_Mri)
Naming=[Glioma_Mri, rot_Brain_tumor_i]
Naming_lb = ['Brain_tumor_Ori', 'Brain_tumor_Rotated']
Brain_tumor_piy.plot_images(Naming,labels=Naming_lb,img_width=180)

# 3. Cropping

crop_Brain_tumor = Brain_tumoroia.Crop(percent=(0, 0.3))
crop_Brain_tumor_i = crop_Brain_tumor.augment_image(Glioma_Mri)
Naming=[Glioma_Mri, crop_Brain_tumor_i]
Naming_lb = ['Brain_tumor_Ori', 'Brain_tumor_Cropped']
Brain_tumor_piy.plot_images(Naming,labels=Naming_lb,img_width=180)

# 4. shearing

shear_Brain_tumor = Brain_tumoroia.Affine(shear=(-40,40))
shear_Brain_tumor_i=shear_Brain_tumor.augment_image(Glioma_Mri)
Naming=[Glioma_Mri, shear_Brain_tumor_i]
Naming_lb = ['Brain_tumor_Ori', 'Brain_tumor_Shearing']
Brain_tumor_piy.plot_images(Naming,labels=Naming_lb,img_width=180)

# 5. Noisy Image

noise_Brain_tumor=Brain_tumoroia.AdditiveGaussianNoise(10,40)
noise_Brain_tumor_i=noise_Brain_tumor.augment_image(Glioma_Mri)
Naming=[Glioma_Mri, noise_Brain_tumor_i]
Naming_lb = ['Brain_tumor_Ori', 'Brain_tumor_Noisy']
Brain_tumor_piy.plot_images(Naming,labels=Naming_lb,img_width=180)
```

**Printing images in grid format**

```python
# Function to load and display images in a grid format
def Brain_tumor_load(Brain_tumor_fdr, Brain_tumor_num=8):
    Fresh_List = []
    tumor_Files = sorted(Brain_tumoro.listdir(Brain_tumor_fdr))[:Brain_tumor_num]
    for tumor_File in tumor_Files:
        path_of_tumor = Brain_tumoro.path.join(Brain_tumor_fdr, tumor_File)
        try:
            iggg_Brainn = Brain_tumorpil.open(path_of_tumor)
            Fresh_List.append(iggg_Brainn)
        except IOError:
            continue
    return Fresh_List

# Function to display images in a grid with a specified number of columns
def Brain_tumor_grid(IIG, colmn=8):
    Rws = (len(IIG) + colmn - 1) // colmn
    figg_Brain_tumor, axess_Brain_tumor = Brain_tumorOTp.subplots(Rws, colmn, figsize=(15, Rws * 3))
    axess_Brain_tumor = axess_Brain_tumor.flatten()

    for axx_Brain, iggg_Brainn in zip(axess_Brain_tumor, IIG):
        axx_Brain.imshow(iggg_Brainn)
        axx_Brain.axis('off')
    for axx_Brain in axess_Brain_tumor[len(IIG):]:
        axx_Brain.axis('off')

    Brain_tumorOTp.show()

    # Display images for each tumor category

def Brain_tumor_display(Route_of_Image, Brain_tumor_cls, img_ct=8):
    for labl in Brain_tumor_cls:
        Brain_tumor_fdr = Brain_tumoro.path.join(Route_of_Image, labl)
        if Brain_tumoro.path.isdir(Brain_tumor_fdr):
            IIG = Brain_tumor_load(Brain_tumor_fdr, Brain_tumor_num=img_ct)
            if IIG:
                print(f'printing images for brain Tumor categpry: {labl}')
                Brain_tumor_grid(IIG)

 # Specify the main directory and the tumor classes

Route_of_Image='/MRI DATA/'

Brain_tumor_cls = ['glioma', 'meningioma', 'notumor', 'pituitary']
Brain_tumor_display(Route_of_Image, Brain_tumor_cls)
```

**Tumor points**

```python
# Display the image with tumor marking
def Brain_tumor_Marking_tumor(Glioma_Mri, Brain_tumor_ps, Brain_tumor_rds, color='red'):
    iggg_Brainn = Brain_tumorpil.open(Glioma_Mri)

    Tumor_draw = Brain_tumorpildr.Draw(iggg_Brainn)

    ## position setting
    xTumor_draw, yTumor_draw = Brain_tumor_ps
    up = (xTumor_draw - Brain_tumor_rds, yTumor_draw - Brain_tumor_rds)
    down = (xTumor_draw + Brain_tumor_rds, yTumor_draw + Brain_tumor_rds)
    Tumor_draw.ellipse([up, down], outline=color, width=2)

    return iggg_Brainn

def Brain_tumor_show(iggg_Brainn):
    Brain_tumorOTp.imshow(iggg_Brainn)
    Brain_tumorOTp.axis('off')
    Brain_tumorOTp.show()

 # Marking and displaying tumor positions on sample images

Glioma_Mri = Route_of_Image+"glioma/Te-gl_0010.jpg"
Brain_tumorOTp.title("glioma Tumor position")
Brain_tumor_ps = (120, 100)
Brain_tumor_mark = Brain_tumor_Marking_tumor(Glioma_Mri, Brain_tumor_ps, 20)
Brain_tumor_show(Brain_tumor_mark)

Meningioma_Mri = Route_of_Image+"meningioma/Te-me_0014.jpg"
Brain_tumorOTp.title("meningioma Tumor position")
Brain_tumor_ps = (400, 300)
Brain_tumor_mark = Brain_tumor_Marking_tumor(Meningioma_Mri, Brain_tumor_ps, 60)
Brain_tumor_show(Brain_tumor_mark)

pituitary_Mri = Route_of_Image+"pituitary/Te-pi_0018.jpg"
Brain_tumorOTp.title("pituitary Tumor position")
Brain_tumor_ps = (250, 200)
Brain_tumor_mark = Brain_tumor_Marking_tumor(pituitary_Mri, Brain_tumor_ps, 30)
Brain_tumor_show(Brain_tumor_mark)
```
